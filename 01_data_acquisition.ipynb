```python
# Uganda Multi-Disease GeoAI Project

import pandas as pd
import geopandas as gpd
import xarray as xr
import matplotlib.pyplot as plt
import os
import warnings
warnings.filterwarnings("ignore")   

print(" Libraries imported successfully for Uganda project!\n")

# ====================== PROJECT PATHS ======================
BASE_DIR = r"I:\Data Science Projects\disease_prediction"
DATA_RAW = os.path.join(BASE_DIR, "data", "raw")
print(f"Raw data directory: {DATA_RAW}\n")

# ====================== ADMINISTRATIVE BOUNDARIES ======================
boundaries_folder = os.path.join(DATA_RAW, "boundaries")
main_shp_path = os.path.join(boundaries_folder, "uga_admin_boundaries.shp")

print("=== Administrative Boundaries ===")
print(f"File exists: {os.path.exists(main_shp_path)}\n")

# Load the most useful layers one by one
layers_to_try = ["uga_admin2", "uga_adm2", "uga_admin1", "uga_adm1", "uga_admin0"]

for layer_name in layers_to_try:
    try:
        gdf = gpd.read_file(main_shp_path, layer=layer_name)
        print(f" Successfully loaded layer: **{layer_name}**")
        print(f"   → Number of features: {len(gdf)}")
        print(f"   → Columns: {list(gdf.columns)}")
        
        # Print sample names if available
        name_cols = [col for col in gdf.columns if any(x in col.lower() for x in ['name', 'en', 'adm'])]
        if name_cols:
            print(f"   → Sample names from column '{name_cols[0]}':")
            print(gdf[name_cols[0]].head(8).tolist())
        print("-" * 60)
        break   
    except Exception as e:
        print(f"   Layer '{layer_name}' failed: {e}")

else:
    print(" Could not load any layer. Please check the file.")

# ====================== POPULATION RASTER ======================
pop_folder = os.path.join(DATA_RAW, "population")
pop_path = os.path.join(pop_folder, "uga_ppp_2020_UNadj_constrained.tif")

print("\n=== Population Raster ===")
print(f"Exists: {os.path.exists(pop_path)}")
if not os.path.exists(pop_path):
    print("→ Download link reminder: https://hub.worldpop.org/geodata/summary?id=49721")
    print("→ Place the .tif file inside data/raw/population/ folder")

# ====================== CDS API ======================
import cdsapi
c = cdsapi.Client()
print("\n CDS API client initialized successfully!")
```
